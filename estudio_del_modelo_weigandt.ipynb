## **Clima por región**
En el presente gráfico boxplot realizado con la librería Seaborn se busca realizar una comparación de temperaturas entre las distintas regiones de españa a fin de tener en cuenta dicho comportamiento al estudiar el consumo de cada una de ellas.

```python
my_palet = sns.color_palette("pastel")
sns.boxplot(x = df_clima_espana['city_name'], y = df_clima_espana['temp_max'], palette = my_palet)
plt.title('Boxplot de temperaturas máximas por región')
plt.show()
```

Como podemos ver el gráfico muestra claramente las zonas más cálidas y las más frías, como así también su amplitud térmica. Así podemos concluir que Sevilla tiene el promedio de temperaturas más altas durante el año, como así también la mayor amplitud térmica, junto a Madrid, solo que este último con temperaturas más templadas.
